```python
# evaluate an lasso regression model on the dataset
from numpy import mean
from numpy import std
from numpy import absolute
from pandas import read_csv
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Lasso
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = read_csv('Data_Problem_3_Sai_Thatigotla.csv')
dataframe = dataframe.fillna(method='bfill')
```

```python
dataframe
```

```python
data = dataframe[['Bedrooms', 'Bathrooms', 'sqft', 'cars_garage', 'stories', 'lot_sqft', 'fireplace', 'kitchen_dim']].values
response = dataframe[['Price']].values
```

```python
# define model
model = Lasso(alpha=1.0, normalize=True)
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, data, response, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force scores to be positive
print(scores)
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
model.fit(data, response)
print(model.coef_)
```
